```python
import pandas as pd
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
import os
```

```python
os.getcwd()
```

```python
train = pd.read_csv("/Users/azizhazeinita/MScA/Winter Quarter/Data Mining Principles/diabetes_train.csv", sep = ",")
test = pd.read_csv("/Users/azizhazeinita/MScA/Winter Quarter/Data Mining Principles/diabetes_test.csv", sep = ",")
```

```python
train.info()
test.info()
```

```python
train = train.drop("Unnamed: 0", 1)
test = test.drop("Unnamed: 0", 1)
```

```python
train.info()
test.info()
```

```python
train.head()
```

```python
test.head()
```

```python
X_train = train.drop(['readmitted'], axis=1)
y_train = train.readmitted
X_test = test.drop(['readmitted'], axis=1)
y_test = test.readmitted
```

```python
#Doing SVM

svm = SVC(kernel='rbf', random_state=42)
svm.fit(X_train, y_train)
```

```python
#Doing Prediction

svm_pred_train = svm.predict(X_train)
svm_pred_test = svm.predict(X_test)
```

```python
print("Confussion Matrix of SVM Train Set: \n", confusion_matrix(y_train, svm_pred_train))
print("\n")
print("Confussion Matrix of SVM Test Set: \n", confusion_matrix(y_test, svm_pred_test))
```

```python
print("Accuracy Score of SVM Train Set: ", accuracy_score(y_train, svm_pred_train))
print("Accuracy Score of SVM Test Set: ", accuracy_score(y_test, svm_pred_test))
```

```python
#Download dataset
df1 = pd.DataFrame(svm_pred_train)
df2 = pd.DataFrame(svm_pred_test)

df1.to_csv('svm_pred_train.csv')
df2.to_csv('svm_pred_test.csv')
```
